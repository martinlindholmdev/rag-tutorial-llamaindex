```python
import os
from llama_index.core import VectorStoreIndex, SimpleDirectoryReader
```

```python
# Set environment variable, this key you need to create at
os.environ["OPENAI_API_KEY"] = "INSERT_YOUR_KEY_HERE"
```

```python
# Read daya from 'data' folder
documents = SimpleDirectoryReader("data").load_data()
```

```python
!pip install EbookLib html2text
```

```python
# Create index and embed documents
index = VectorStoreIndex.from_documents(documents)
```

```python
# Create query engine
query_engine = index.as_query_engine()
```

```python
# Query the database
response = query_engine.query("Vad menar författarna med 'kaospusslet? Svara på svenska.")
print(response)
```
